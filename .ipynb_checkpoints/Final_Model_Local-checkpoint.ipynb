```python
class ResBlock(tf.keras.layers.Layer):
    def __init__(self, filters=32, enlarge=8, shrink=0.8):
        super(ResBlock, self).__init__()
        self.initializer = tf.keras.initializers.GlorotNormal()
        #self.regulizer = tf.keras.regularizers.L1L2(l1=0.01, l2=0.01)
        self.regulizer = NoneW+TZl5h{gsMU
        self.Conv1 = layers.Conv2D(filters=filters * enlarge, 
                                   kernel_size=1, 
                                   padding='same', 
                                   kernel_initializer=self.initializer, 
                                   kernel_regularizer=self.regulizer)
        self.Act = tf.keras.activations.swish
        self.Conv2 = layers.Conv2D(filters=int(filters * shrink), kernel_size=1, padding='same', kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.Conv3 = layers.Conv2D(filters=filters, kernel_size=3, padding='same', kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.Add = tf.keras.layers.Add()

    def call(self, inputs):
        x = self.Conv1(inputs)
        x = self.Act(x)
        x = self.Conv2(x)
        x = self.Conv3(x)
        return self.Add([inputs, x])
```

```python
class MainRes(tf.keras.layers.Layer):
    def __init__(self, filters=32, enlarge=8, shrink=0.8):
        super(MainRes, self).__init__()
        self.initializer = tf.keras.initializers.GlorotNormal()
        #self.regulizer = tf.keras.regularizers.L1L2(l1=0.01, l2=0.01)
        self.regulizer = None
        self.SRMain1 = layers.Conv2D(filters=32, kernel_size=3, padding="same", kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.SRMainLoop = [ ResBlock() for i in range(0, 8) ]
        self.SRMain2 = layers.Conv2D(filters=48, kernel_size=3, padding="same", kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.SRMainDepthToSpace = tf.keras.layers.Lambda(lambda x: tf.nn.depth_to_space(x, 4))

    def call(self, inputs):
        x = self.SRMain1(inputs)
        for block in self.SRMainLoop:
            x = block(x)
        x = self.SRMain2(x)
        return self.SRMainDepthToSpace(x)
```

```python
class SkipRes(tf.keras.layers.Layer):
    def __init__(self):
        super(SkipRes, self).__init__()
        self.initializer = tf.keras.initializers.GlorotNormal()
        #self.regulizer = tf.keras.regularizers.L1L2(l1=0.01, l2=0.01)
        self.regulizer = None
        self.SRSkip1 = layers.Conv2D(filters=48, kernel_size=5, padding="same", kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.SRSkipDepthToSpace = tf.keras.layers.Lambda(lambda x: tf.nn.depth_to_space(x, 4))

    def call(self, inputs):
        x = self.SRSkip1(inputs)
        return self.SRSkipDepthToSpace(x)
```

```python
class WDSR(tf.keras.layers.Layer):
    def __init__(self, mse_loss_rate=0.5):
        super(WDSR, self).__init__()
        self.SRAdd = tf.keras.layers.Add()
        self.Main = MainRes()
        self.Skip = SkipRes()
        self.MSERate = mse_loss_rate
        self.MSE = tf.keras.losses.MeanSquaredError()
        self.Resize = tf.keras.layers.Resizing(224, 224, interpolation="bilinear")

    def psnr(self, original, lowres):
        return -10 * K.log(K.mean(K.flatten((original - lowres)) ** 2)) / np.log(10)

    def ssim(self, original, lowres):
        max_val = tf.math.reduce_max([tf.math.reduce_max(original), tf.math.reduce_max(lowres)])
        min_val = tf.math.reduce_min([tf.math.reduce_min(original), tf.math.reduce_min(lowres)])
        return tf.image.ssim(original, lowres, max_val - min_val)

    def get_config(self):
        config = super().get_config()
        config.update({
            "mse_loss_rate": self.MSERate,
        })
        return config

    def call(self, inputs, training=False):
        original = inputs
        lowres = self.Resize(inputs)
        m = self.Main(lowres)
        s = self.Skip(lowres)
        sr = self.SRAdd([m, s])

        self.add_loss(self.MSERate * self.MSE(original, sr))

        self.add_metric(self.psnr(original, sr), 'psnr')
        self.add_metric(self.ssim(original, sr), 'ssim')

        return sr
```

```python
class TransferClassifier(tf.keras.layers.Layer):
    def __init__(self, size):
        super(TransferClassifier, self).__init__()
        self.size = size
        self.initializer = tf.keras.initializers.GlorotNormal()
        #self.regulizer = tf.keras.regularizers.L1L2(l1=0.01, l2=0.01)
        self.regulizer = None
        # Transfer Model
        
        self.preprocessor = tf.keras.applications.resnet50.preprocess_input
        self.transfer = tf.keras.applications.ResNet50V2(
            input_shape=(size[0], size[1], size[2]),
            include_top=False,
            weights='imagenet',
            pooling='avg'
        )
        self.transfer.trainable = False

        # Head
        self.Dense1 = layers.Dense(1024, activation='swish', kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.Dense2 = layers.Dense(512, activation='swish', kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.Dense3 = layers.Dense(256, activation='swish', kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.Dropout1 = layers.Dropout(0.5)
        self.Dense4 = layers.Dense(128, activation='swish', kernel_initializer=self.initializer, kernel_regularizer=self.regulizer)
        self.Dropout2 = layers.Dropout(0.3)
        self.Out = layers.Dense(1, activation='sigmoid')

    def get_config(self):
        config = super().get_config()
        config.update({
            "size": self.size
        })
        return config

        
    def call(self, inputs, training=False):
        x = self.preprocessor(inputs)
        x = self.transfer(x)

        x = self.Dense1(x)
        x = self.Dense2(x)
        x = self.Dense3(x)
        if training:
            x = self.Dropout1(x)
        x = self.Dense4(x)
        if training:
            x = self.Dropout2(x)
        return self.Out(x)
```

```python
class SRLoss(tf.keras.layers.Layer):
    def __init__(self, mse_loss_rate):
        super(SRLoss, self).__init__()
        self.MSERate = mse_loss_rate
        self.MSE = tf.keras.losses.MeanSquaredError()

    def get_config(self):
        config = super().get_config()
        config.update({
            "mse_loss_rate": self.MSERate,
        })
        return config
        
    def psnr(self, original, lowres):
        return tf.image.psnr(original, lowres, 1)

    def ssim(self, original, lowres):
        max_val = tf.math.reduce_max([tf.math.reduce_max(original), tf.math.reduce_max(lowres)])
        min_val = tf.math.reduce_min([tf.math.reduce_min(original), tf.math.reduce_min(lowres)])
        return tf.image.ssim(original, lowres, max_val - min_val)

    def call(self, inputs):
        original = inputs[0]
        sr = inputs[1]
        self.add_loss(self.MSERate * self.MSE(original, sr))
        self.add_metric(self.psnr(original, sr), 'psnr')
        self.add_metric(self.ssim(original, sr), 'ssim')
        self.add_metric(self.MSE(original, sr), 'MSE')
        return sr
```

```python
cbis_ddsm_mean = 159.44312806051445
#cbis_ddsm_var = 1811.82013
#cbis_ddsm_std = 42.5654805
cbis_ddsm_std = 127.5
```

```python
def get_wdsr():
    def weightnorm_conv2d(filters, kernel_size, padding='same', activation=None, kernel_initializer=None, kernel_regularizer=None, **kwargs):
        return tfp.layers.weight_norm.WeightNorm(layers.Conv2D(filters, 
                                                               kernel_size, 
                                                               padding=padding,
                                                               activation=activation, 
                                                               kernel_initializer=kernel_initializer,
                                                               kernel_regularizer=kernel_regularizer,
                                                               **kwargs), data_init=False)
    initializer = tf.keras.initializers.GlorotNormal()
    regulizer = None
    shrink = 0.8
    enlarge = 8

    inputs = tf.keras.Input(shape=(None, None, 3))
    lowres = tf.keras.layers.Resizing(224, 224, interpolation="bilinear")(inputs)
    lowres = tf.keras.layers.Normalization(mean=cbis_ddsm_mean, variance=cbis_ddsm_std**2)(lowres)

    #WDSR
    ##Main Res
    m = weightnorm_conv2d(filters=32, 
                      kernel_size=3, 
                      padding="same", 
                      kernel_initializer=initializer, 
                      kernel_regularizer=regulizer)(lowres)

    for i in range(0, 8):
        m_in = weightnorm_conv2d(filters=192, 
                                   kernel_size=1, 
                                   padding='same', 
                                   kernel_initializer=initializer, 
                                   kernel_regularizer=regulizer, 
                                 activation='relu')(m)
        m_in = weightnorm_conv2d(filters=int(32 * shrink), 
                             kernel_size=1, 
                             padding='same', 
                             kernel_initializer=initializer, 
                             kernel_regularizer=regulizer)(m_in)
        m_in = weightnorm_conv2d(filters=32, 
                             kernel_size=3, 
                             padding='same', 
                             kernel_initializer=initializer, 
                             kernel_regularizer=regulizer)(m_in)
        m = tf.keras.layers.Add()([m, m_in])

    m = weightnorm_conv2d(filters=48, 
                      kernel_size=3, 
                      padding="same", 
                      kernel_initializer=initializer, 
                      kernel_regularizer=regulizer)(m)
    m = tf.keras.layers.Lambda(lambda x: tf.nn.depth_to_space(x, 4))(m)

    ##Skip Res
    s = weightnorm_conv2d(filters=48, 
                      kernel_size=5, 
                      padding="same", 
                      kernel_initializer=initializer, 
                      kernel_regularizer=regulizer)(lowres)
    s = tf.keras.layers.Lambda(lambda x: tf.nn.depth_to_space(x, 4))(s)

    sr = tf.keras.layers.Add()([m, s])
    sr = sr * cbis_ddsm_std
    out = sr + cbis_ddsm_mean
    return tf.keras.Model(inputs=inputs, outputs=out, name="wdsr")
```

```python
def my_psnr(y_true, y_pred):
    return tf.image.psnr(y_true, y_pred, 255.)
def my_ssim(y_true, y_pred):
    return tf.image.ssim(y_true, y_pred, 255.)
```

```python
def get_callbacks(name: str = None):
  if name:
    log_dir = "logs/fit/" + name
  else:
    log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
  tensorboard_callback = tf.keras.callbacks.TensorBoard(
      log_dir=log_dir, 
      histogram_freq=1,
      profile_batch=(0,20)
  )

  check_point = tf.keras.callbacks.ModelCheckpoint(
      filepath="checkpoint.h5",
      monitor="val_loss",
      mode="max",
      save_best_only=True,
  )
  
  schedule = tf.keras.callbacks.LearningRateScheduler(
      schedule= lambda x,y : y / 2 if (x+1) % 2e5 == 0 else y,
      verbose=1
  )
  reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.1,
    patience=3,
    verbose=0,
    mode="auto",
    min_delta=0.0001,
    cooldown=0,
    min_lr=0,
  )

  return [tensorboard_callback, check_point, DisplayProgress(), reduce_lr]
```

```python
def get_model():
    initializer = tf.keras.initializers.GlorotNormal()
    regulizer = None
    enlarge = 8
    shrink = 0.8

    inputs = tf.keras.Input(shape=(224, 224, 3))
    inputs = tf.keras.layers.Normalization(mean=cbis_ddsm_mean, variance=cbis_ddsm_std**2)(inputs)
    lowres = tf.keras.layers.Resizing(224, 224, interpolation="bilinear")(inputs)

    #WDSR
    wdsr.trainable=True
    sr = wdsr(lowres)
    #sr = SRLoss(1)([inputs, sr])

    #Transfer Classifier
    downsuperres = tf.keras.layers.Resizing(224, 224, interpolation="mitchellcubic")(sr)
    transfer = tf.keras.applications.ResNet50V2(
            input_shape=(224, 224, 3),
            include_top=False,
            weights='imagenet',
            pooling='avg'
        )
    transfer.trainable=False
    x = transfer(downsuperres)
    x = layers.Dense(1024, 
                     activation='swish', 
                     kernel_initializer=initializer, 
                     kernel_regularizer=regulizer)(x)
    x = layers.Dense(512, 
                     activation='swish', 
                     kernel_initializer=initializer, 
                     kernel_regularizer=regulizer)(x)
    x = layers.Dense(256, 
                     activation='swish', 
                     kernel_initializer=initializer, 
                     kernel_regularizer=regulizer)(x)
    x = layers.Dense(128, 
                     activation='swish', 
                     kernel_initializer=initializer, 
                     kernel_regularizer=regulizer)(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    return tf.keras.Model(inputs=inputs, outputs=out, name="combined_model")
```

```python
def get_callbacks(name: str = None):
  if name:
    log_dir = "logs/fit/" + name
  else:
    log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
  tensorboard_callback = tf.keras.callbacks.TensorBoard(
      log_dir=log_dir, 
      histogram_freq=1,
      profile_batch=(0,20)
  )

  check_point = tf.keras.callbacks.ModelCheckpoint(
      filepath="checkpoint.h5",
      monitor="val_loss",
      mode="max",
      save_best_only=True,
  )

  reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.1,
    patience=3,
    verbose=0,
    mode="auto",
    min_delta=0.0001,
    cooldown=0,
    min_lr=0,
  )
  
  schedule = tf.keras.callbacks.LearningRateScheduler(
      schedule= lambda x,y : y / 5 if (x+1) % 5 == 0 else y,
      verbose=1
  )

  return [tensorboard_callback, check_point, reduce_lr]
```

```python
def apply_clahe(image_path: str, size: int, size_2: int = -1) -> np.array:
    image = cv2.imread(image_path)
    if size < 0:
        pass
    elif size_2 > 0:
        image = cv2.resize(image, (size, size_2))
    else:
        image = cv2.resize(image, (size, size))
    image_bw = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    clahe = cv2.createCLAHE(clipLimit = 5)
    return np.repeat(clahe.apply(image_bw)[:, :, np.newaxis], 3, axis=2).astype('uint8')
```
